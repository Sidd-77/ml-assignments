```python
#4. Display count of rows having null in any column
null_counts_per_row = df.isnull().sum(axis=1)
rows_with_null_values = df[null_counts_per_row > 0]
count_rows_with_null_values = len(rows_with_null_values)
print(count_rows_with_null_values)
```

```python
#5. Display first 10 rows
print(df.head(10))
```

```python
#6. Display last 10 rows
print(df.tail(10))
```

```python
#7. Display rows having number of floors more than or equal to 1.5
filtered_rows = df[df['floors'] >= 1.5]
print(filtered_rows)
```

```python
#8. Display houses older than 1968 and their count
older_houses = df[df['yr_built']<1968]
older_houses_count = len(older_houses)
print(older_houses)
print("Count of houses older than 1968 : ", older_houses_count)
```

```python
#9. Display houses having price more than 50000/- and their count
houses_costly = df[df['price']>50000]
houses_costly_count = len(houses_costly)
print(houses_costly)
print("Count of houses which have price more than 50000/- : ",houses_costly_count)
```

```python
#10. Display houses having number of bedrooms less than or equal to 2 and price above 51000/- along with their count
fileterd_rows = df[(df['price']>51000) & (df['bedrooms']<=2)]
fileterd_rows_count = len(fileterd_rows)
print(fileterd_rows)
print("Count of houses having number of bedrooms less than or equal to 2 and price above 51000/- : ", fileterd_rows_count)
```

```python
#11. Display houses having number of bedrooms more than 2 and square foot
fileterd_rows2 = df[df['bedrooms']>2]
print(fileterd_rows2)
```

```python
#12. Replace/eliminate missing values

#Drops rows with null values
df.dropna(inplace=True)
```

```python
#13. Change column name(s) to short/easy names if required.

df.rename(columns={'sqft_basement': 'sqft_bsmt'}, inplace=True)
print(df.columns)
```

```python
#14. Drop unessential columns (feature selection).

df.drop('sqft_living15', axis=1, inplace=True)
print(df.columns)
```

```python
#15. Find mean/min/max of numeric columns.

mean_value = df.mean()
min_value = df.min()
max_value = df.max()

print("Mean values : ", mean_value)
print("Min values : ", min_value)
print("Max values : ", max_value)
```

```python
#16. Find mode of all columns.
mode_value = df.mode()
print("Mode values : ", mode_value)
```

```python
#17. Display unique values in each column

for column in df.columns :
  unique_values = df[column].unique()
  print("Unique values in ", column)
  print(unique_values)
  print("\n")
```
