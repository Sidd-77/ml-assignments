```python
dummy_df = pd.read_csv('./hungary_chickenpox.csv')

# Define the proportion of nulls you want to introduce (e.g., 5%)
prop_nulls = 0.05

# Calculate the number of nulls to introduce
num_nulls = int(prop_nulls * dummy_df.size)

# Generate random indices for rows
row_indices = np.random.randint(low=0, high=dummy_df.shape[0], size=num_nulls)

# Generate random indices for columns
col_indices = np.random.randint(low=0, high=dummy_df.shape[1], size=num_nulls)

# Assign np.nan to the random indices
for i in range(num_nulls):
    dummy_df.iat[row_indices[i], col_indices[i]] = np.nan

print("Number of null values in dataset : ")
print(dummy_df.isnull().sum().sum())
```

```python
# Determine percentage of null values column-wise
print(dummy_df.isnull().sum() / dummy_df.shape[0] * 100)
```

```python
# Performing Imputation on dummy dataframe using mean and mode method

# For numeric columns, fill in the missing values with the mean
for col in dummy_df.select_dtypes(include=[np.number]).columns:
    dummy_df[col].fillna(dummy_df[col].mean(), inplace=True)

# For non-numeric columns, fill in the missing values with the mode
for col in dummy_df.select_dtypes(exclude=[np.number]).columns:
    dummy_df[col].fillna(dummy_df[col].mode().iloc[0], inplace=True)
```

```python
print("After Imputaion of missing values : ")
print(dummy_df.isnull().sum().sum())
```

```python
#Standardization
numeric_df_standardized = (numeric_df - numeric_df.mean()) / numeric_df.std()

#Normalization
numeric_df_normalized = (numeric_df - numeric_df.min()) / (numeric_df.max() - numeric_df.min())

# Visualize standardization
for col in numeric_df_standardized.columns:
    plt.figure()
    plt.hist(numeric_df_standardized[col].dropna(), bins=30)  # dropna to remove any NaN values
    plt.title(f'Standardized {col}')
    plt.show()

# Visualize normalization
for col in numeric_df_normalized.columns:
    plt.figure()
    plt.hist(numeric_df_normalized[col].dropna(), bins=30)  # dropna to remove any NaN values
    plt.title(f'Normalized {col}')
    plt.show()
```

```python

arr = [10,50,100,500,1000,1500]

import timeit
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

m_values = [10,50,100,500,1000,1500]
n = 10000
pca_times = []
tsne_times = []

for m in m_values:
    # Create a dataset with n data points and m features
    dataset = np.random.randn(n, m)

    # Measure the execution time for PCA
    start_time = timeit.default_timer()
    pca = PCA(n_components=2)
    pca.fit_transform(dataset)
    end_time = timeit.default_timer()
    pca_times.append(end_time - start_time)

    # Measure the execution time for t-SNE
    start_time = timeit.default_timer()
    tsne = TSNE(n_components=2)
    tsne.fit_transform(dataset)
    end_time = timeit.default_timer()
    tsne_times.append(end_time - start_time)

# Plot m vs execution time
plt.plot(m_values, pca_times, label='PCA')
plt.plot(m_values, tsne_times, label='t-SNE')
plt.xlabel('m (number of features)')
plt.ylabel('Execution time (seconds)')
plt.legend()
plt.show()
```
